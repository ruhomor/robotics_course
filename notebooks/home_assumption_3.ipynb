# Сколько пальцев?

Напришите программу, которая будет в режиме реального времени выводить, сколько пальцев Вы показываете в камеру.

Детектор цветного объекта из первого задания можно использовать для получения маски руки.

Затем можно проанализировать маску, например с помощью функции skeletonize из второго семинара, получив тонкий силуэт руки, и посчитать количество концов линий.

Количество пальцев можно выводить просто цифрой в углу экрана.

P.S. Стабильно выделяется рука без шума - 1 балл. При увеличении количества пальцев в кадре увеличивается выводимое значение - 4 балла. Правильно выводится количество пальцев, когда их от 1 до 5 - 7 баллов. Правильно обрабатывается случай 0 пальцев - 9 баллов. Программа работает с fps больше 3 -- +1 балл.
